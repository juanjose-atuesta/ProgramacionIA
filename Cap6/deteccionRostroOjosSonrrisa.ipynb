```python
import cv2
img = cv2.imread('content/pexels-alex.jpg')
faseClassifier = cv2.CascadeClassifier('content/haarcascade_frontalface_default.xml')#el archivo xml tiene el modelo ya entrenado
eyeClassifier = cv2.CascadeClassifier('content/haarcascade_eye.xml')
smileClassifier = cv2.CascadeClassifier('content/haarcascade_smile.xml')
imgGrey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faseClassifier.detectMultiScale(imgGrey, scaleFactor=1.3, minNeighbors=5, minSize=(30,30),)
#Funcion para obtener el Ã¡rea del rostro detectado  
def getFaceAreaDetected(faces):
    for (x, y, w, h) in faces:
        return imgGrey[y:y+h, x:x+w]
def getFaceAreaColorDetected(faces):
    for (x, y, w, h) in faces:
        return img[y:y+h, x:x+w]

for (x, y, w, h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    cv2.putText(img,'face',(x+10,y+10),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,0,0),2)
    eyes = eyeClassifier.detectMultiScale(getFaceAreaDetected(faces), scaleFactor=1.2, minNeighbors=15, minSize=(2,2),)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(getFaceAreaColorDetected(faces),(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        cv2.putText(img,'eye',(x+ex+10,y+ey+10),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,0),2)
        smiles = smileClassifier.detectMultiScale(getFaceAreaDetected(faces), scaleFactor=1.2, minNeighbors=22, minSize=(20,20),)
        for (sx, sy, sw, sh) in smiles:
            cv2.rectangle(getFaceAreaColorDetected(faces),(sx,sy),(sx+sw,sy+sh),(0,0,255),2)
            cv2.putText(img,'smile',(x+sx+10,y+sy+10),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),2)
    
cv2.imshow('img', img)
cv2.waitKey(0)
```
