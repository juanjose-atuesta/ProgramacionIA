```python
"""Ejemplo 6.1: Detección de rostros usando la librería opencv y el detector
   de rostros haar (faceharrcascade.py)
Se usa el modelo Haar Cascade Classifier"""

""" Este modelo es un modelo de clasificacion que detecta rostros, no los reconoce. Responde a la pregunta: hay un rostro aqui?
    lo que hace el modelo es ver contrastes de luz, por ejemplo:
    -Los ojos suelen ser mas oscuros que la frente o las mejillas
    -el puente de la nariz es mas claro
    -la frente suele ser mas clara
    EL modelo usa rectangulos blancos y negros para medis diferencias, eso se llama caracteristicas Haar:
    ⬜⬛⬜   ← patrón típico de ojos y nariz
    Lo que hace el modelo es recorrer la imagen en una centana, aplicando pruebas simples sobre esa ventana y se pregunta algo como Esta parte es mas clara que esta otra?
    si muchas pruebas coenciden, es posible que haya un rostro.
    Que significa "cascade"? EL modelo identifica a travez de diferentes etapas, en la que cada vez mas se reviza con mas presicion y detalle la imagen, eliminando
    primeramente todo lo que definitivamente no es un posible rostro. AL final confirma  si es un rostro. Si falla en una etapa, se descarta inmediatamente.
    Por ultimo aclarar que cascade_classifier lo que retorna es una lista de rectangulos donde cree que hay un rostro, de la forma (x, y, w, h)
     donde (x,y) es la esquina superior izquierda, W y H representan el ancho y alto respecticamente 
"""
import cv2

#Carga la imagen de prueba de un directorio de de Google Colaboratory
img = cv2.imread('content/pexels-pixabay.jpg')

#Carga el detector de rostros Haarcascade
cascade_classifier = cv2.CascadeClassifier('content/haarcascade_frontalface_default.xml')#el archivo xml tiene el modelo ya entrenado
#cascade_classifier = cv2.CascadeClassifier('content/haarcascade_frontalface_default.xml')

#Convierte la imagen de prueba a una escala de grises
gray =cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#Identifica las coordenadas del rostro con ayuda del detector Haar
faces= cascade_classifier.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

#Ciclo for para trazar el rectángulo delimitador del rostro
for (x,y,w,h) in faces:
        # Traza el rectángulo delimitador del rostro y le pone formato al rectángulo
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        # Escribe la palabra face encima del rectángulo
        cv2.putText(img,'face', (x + 10, y + 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,255), 2)

#Muestra la imagen con el rectángulo trazado
#cv2.imshow('face', img)
cv2.imshow('face', img)
cv2.waitKey(0)
```
