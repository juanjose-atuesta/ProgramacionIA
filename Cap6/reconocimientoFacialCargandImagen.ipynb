```python
import pathlib
import face_recognition
import os
import numpy as np
from PIL import Image, ImageDraw
import cv2
path = "/home/juan-jose/Documentos/Codigos/Personal/ProgramaciÃ³nDeInteligenciaArtificial/Cap6/content/"

def getFacesOfTheContentDir(path):
    imagespaths= [os.path.join(path,f) for f in os.listdir(path)]#crea una lista con las rutas de las imagenes contenidas en el path
    facesCodings = []
    facesIDs = []
    for imagepath in imagespaths:
         if pathlib.Path(os.path.split(imagepath)[1]).suffix == ".jpg" or pathlib.Path(os.path.split(imagepath)[1]).suffix == ".jpeg":
                image = face_recognition.load_image_file(imagepath)
                face_encoding = face_recognition.face_encodings(image)[0]
                facesCodings.append(face_encoding)
                ID = os.path.split(imagepath)[1].split('.')[0]#obtiene el nombre del archivo
                facesIDs.append(ID)
    return facesIDs, facesCodings

knowkFacesIDs, knowkFacesEncodings = getFacesOfTheContentDir(path)
```

```python
unknownFile = face_recognition.load_image_file("content/mari2.jpeg")
imageFile = cv2.imread('content/mari2.jpeg')
unknownFace = face_recognition.face_locations(unknownFile)
encodingFaceTest = face_recognition.face_encodings(unknownFile,unknownFace)[0]
pilImage = Image.fromarray(unknownFile)
draw = ImageDraw.Draw(pilImage)
#print(knowkFacesEncodings.shape)
matches = face_recognition.compare_faces(knowkFacesEncodings,encodingFaceTest)
print(matches)
```

```python
for (top,right, bottom, left) in unknownFace:
    matches = face_recognition.compare_faces(knowkFacesEncodings,encodingFaceTest)
    distances = face_recognition.face_distance(knowkFacesEncodings,encodingFaceTest)

    indexMin = np.argmin(distances)
    name = knowkFacesIDs[indexMin]
    print(name)
    # Draw a box around the face
    cv2.rectangle(imageFile, (left, top), (right, bottom), (0, 0, 255), 2)

    # Draw a label with a name below the face
    cv2.rectangle(imageFile, (left, bottom - 35), (right, bottom), (0, 0, 255), cv2.FILLED)
    font = cv2.FONT_HERSHEY_DUPLEX
    cv2.putText(imageFile, name, (left + 6, bottom - 6), font, 1.0, (255, 255, 255), 1)

# Display the resulting image
cv2.imshow('Image', imageFile)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
