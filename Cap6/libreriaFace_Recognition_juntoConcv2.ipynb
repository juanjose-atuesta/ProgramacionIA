```python
"""Deteccion de rostros usando la biblioteca Face_recognition. Esta librerias es mas simple y exacta al momento de reconocer rostros"""
import face_recognition
from PIL import Image
import cv2
```

```python

#Carga la imagen de prueba
image = face_recognition.load_image_file('content/mari.jpeg')

#Identifica las coordenadas del rostro de prueba
face_locations = face_recognition.face_locations(image)

#Imprime las coordenadas del rostro de prueba
count = 0
for face_location in face_locations:
      count = count + 1
      top, right, bottom, left = face_location
      print("Face {} Top: {}, Left: {}, Bottom: {}, Right: {}".format(count,
                                                top, left, bottom, right))

#Imprime la imagen con los respectivos límites del rostro
face_image = image[top:bottom, left:right]
#pil_image = Image.fromarray(face_image)
#pil_image.show()
#no entendi porque no funciona, pero con cv2 es mejor
face_image_bgr = cv2.cvtColor(face_image, cv2.COLOR_RGB2BGR)
cv2.imshow("Face detected", face_image_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
"""Con cv2, la misma monda que arriba"""

#Carga la imagen, la muestra y la convierte a formato RGB
image = cv2.imread("content/pexels-mateus.jpg")
cv2.imshow('photo', image)
rgb_frame = image[:, :, ::-1]

#Identifica las coordenadas del rostro de prueba
face_locations = face_recognition.face_locations(rgb_frame)

#Imprime las coordenadas del rostro de prueba
count = 0
for face_location in face_locations:
        count = count + 1
        top, right, bottom, left = face_location
        print("Face {} Top: {}, Left: {}, Bottom: {}, Right: {}".format(count, top, left, bottom, right))

#Imprime la imagen con los respectivos límites del rostro
face_image = image[top:bottom, left:right]
title = 'face' + str(count)
cv2.imshow(title, face_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
